```python
aapl=close.AAPL.copy().to_frame()
aapl
```

```python
aapl.shift(periods=1)
```

```python
aapl["lag1"]=aapl.shift(periods=1)
```

```python
aapl
```

```python
aapl["diff"]=aapl.AAPL.sub(aapl.lag1)
aapl
```

```python
aapl["%change"]=aapl.AAPL.div(aapl.lag1).sub(1).mul(100)
aapl
```

```python
# del aapl["diff"]
```

```python
aapl=aapl.rename(columns={"%_change":"%change"})
```

```python
aapl
```

```python
# resampling data

aapl.AAPL.resample("ME").last()
```

```python
aapl.AAPL.resample("BME").last().pct_change(periods=1).mul(100)#bm=business day;m=last day of month
```

```python
# del aapl["diff"]
del aapl["lag1"]
del aapl["%change"]
```

```python
aapl
```

```python
ret=aapl.pct_change().dropna()
ret
```

```python
# calculte moving average:
spy=yf.download("SPY")
```

```python
spy=spy.Close
spy
```

```python
spy
```

```python
spy.plot(figsize=(12,8),fontsize=15)
plt.legend(loc="upper left" , fontsize=15)
plt.show()
```

```python
spy_roll=spy.rolling(window=10) 
# rolling function calculate the past experience:
spy_roll
```

```python
spy_roll=spy_roll.mean()
```

```python
spy_roll.mean().head(15)
```

```python
spy_roll.rolling(window=10).max()
```

```python
spy.rolling(window=10,min_periods=5).max()
```

```python
spy.rolling(window=10,min_periods=5).max().head(15)
```

```python
spy
```

```python
spy["SMA50"]=spy.rolling(window=10,min_periods=5).mean()
spy
```

```python
spy.head(52)
```

```python
spy.plot(figsize=(12,8),fontsize=15)
plt.legend(loc="upper left" , fontsize=15)
plt.show()
```

```python
spy["SMA200"]=spy.SPY.rolling(window=200,min_periods=200).mean()
spy
```

```python
spy.plot(figsize=(12,8),fontsize=15)
plt.legend(loc="upper left" , fontsize=15)
plt.show()
```
